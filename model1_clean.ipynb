```python
data_to_add = pd.read_csv("model1_wb_data.csv")
data_rel_rows = data_to_add[:56]
# num_of_attrs = data_rel_rows.groupby("Country Code").transform(lambda x:x) 
num_of_attrs = 8



flood_data = pd.read_csv("condensed_model1.csv")
# data_rel_rows
```

```python
# flood_data['Year'] = pd.to_datetime(flood_data['Date of event (start)']).dt.year
# flood_data["Country Code"] = flood_data["ISO3"]

for i in range(0, data_rel_rows.shape[0], num_of_attrs):
    _sub = data_rel_rows[i:i+num_of_attrs] 
    # print(_sub.shape[0])
    name = data_rel_rows["Series Name"][i]
    
    # melt 
    _sub = pd.melt(_sub, id_vars=['Series Name', 'Series Code', 'Country Name', 'Country Code'], var_name='Year', value_name='Value')

    # extract the year from the variable column
    _sub['Year'] = _sub['Year'].str.extract('(\d+)', expand=False).astype(int)

    # display the resulting dataframe
    _portion = _sub[["Year", "Country Code", "Value"]]
    

    # merge the two dataframes on "Country Code" and "Year"
    merged_df = pd.merge(flood_data, _portion[['Country Code', 'Year', 'Value']], on=['Country Code', 'Year'], how='left')

    # rename the "Value" column to "Displaced Population"
    merged_df.rename(columns={'Value': name}, inplace=True)
    flood_data = merged_df 
    
flood_data

```

```python
flood_data.to_csv("model1_cleaned_data.csv")
```
